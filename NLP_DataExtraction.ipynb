```python
import spacy
import nltk
import re
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
urls = [
    "https://en.wikipedia.org/wiki/Napoleon",
    "https://en.wikipedia.org/wiki/Albert_Einstein",
    "https://en.wikipedia.org/wiki/Leonardo_da_Vinci"
]
document=''
# Function to remove text within brackets
def remove_brackets(text):
    return re.sub(r'\[\w+\]', '', text)


for url in urls:
    # Send a GET request to the URL and retrieve the HTML content
    response = requests.get(url)
    html_content = response.text

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, "html.parser")

    # Find all paragraphs on the page
    paragraphs = soup.find_all('p')

    # Concatenate all text from paragraphs
    all_text = "".join(p.text for p in paragraphs)
    print("URL:", url)
    document = document+remove_brackets(all_text)
```

```python
def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = text.lower()
    tokenized_text = text.lower().split()
    tokenized_text = [word for word in tokenized_text if word not in stopwords.words('english')]
    return ' '.join(tokenized_text)
    return text
nlp=spacy.load('en_core_web_lg')
scraped_doc=nlp(document)
tokenized_text_words = [token.text for token in scraped_doc]
tokinized_text_lemma = [token.lemma_ for token in scraped_doc]
tokinized_text_sentences = [sent.text for sent in scraped_doc.sents]
tokinized_text_pos = [(token.text, token.pos_) for token in scraped_doc]
tokinized_text_ner = [(ent.text, ent.label_) for ent in scraped_doc.ents]
unique_words=set(tokenized_text_words)
```
